```python
class CellPhone():
    def __init__(self,model, color, ram,storage,pro,price,camera):
        self.model = model
        self.color = color
        self.ram =ram
        self.storage = storage
        self.pro = pro
        self.price = price
        self.camera = camera
    def spcecs(self):
        print(f"""
                Model:    {self.model}
                Color:    {self.color}
                Ram :     {self.ram}
                Storage:  {self.storage}
                Processor:{self.pro}
                Camera:   {self.camera}
                Price:    {self.price}
        """)
        
    def dial(self, name):
        print(f"Phone is calling {name}")
    def msg(self,name, msg):
        print(f"Phone is sending {msg} to {name}")
        
```

```python
c1 = CellPhone('Iphone','Black','4Gb','64Gb','Snap','100Mp',200000)
```

```python
c1.spcecs()
```

```python
c1.dial('Jhon')
```

```python
c1.msg('Jhon',"where r u ")
```
